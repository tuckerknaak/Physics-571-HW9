```python
#Imports
import matplotlib.pyplot as plt
from numpy import sqrt, linspace


#Constants for NaCl
e = 1.6*10**-19    #charge of electron [C]
M = 2.32*10**-26   #reduced mass [kg]
n = 10**13         #dampening coefficient
NV = 2.24*10**28   #N/V [1/m**3]
e0 = 8.85*10**-12  #permittivity of free space [F/m]
w0 = 4.39*10**13   #resonance frequency [rad/s]
eopt = 2.34        #optical dielectric constant


#Loop to determine values of the dielectric function e(w) and index of refraction N = n +ik
w_range = linspace(10**12,10**14,1000)  #range of frequencies
er_list = []
ei_list = []
emag_list = []
nw_list = []
kw_list = []

for w in w_range:
    er = (((NV*e**2)/(e0*M))*((w0**2-w**2)/((w0**2-w**2)**2+n**2*w**2)))+eopt  #real part of dielectric function
    ei = ((NV*e**2)/(e0*M))*((n*w)/((w0**2-w**2)**2+n**2*w**2))                #imaginary part of dielectric function
    emag = sqrt((er**2)+(ei**2))  #magnitude of dieletric function
    nw = sqrt((emag+er)/2)        #n(w)
    kw = sqrt((emag-er)/2)        #k(w)
    er_list.append(er)
    ei_list.append(ei)
    emag_list.append(emag)
    nw_list.append(nw)
    kw_list.append(kw)


#Plotting the figures
plt.figure(figsize=(10,6),dpi=100)
plt.plot(w_range,nw_list,linestyle='solid',color='dodgerblue',label='n(w)')
plt.plot(w_range,kw_list,linestyle='solid',color='limegreen',label='k(w)')
plt.grid(True,alpha=0.6,linestyle='dashed',color='darkgray',which='both')
plt.legend(prop={'size':15})
plt.rcParams['legend.loc'] = 'upper right'
plt.title('n(w) and k(w) - One Harmonic Oscillator')
plt.xlabel('w [rad/s]')
plt.ylabel('n(w), k(w)')
plt.show()

plt.figure(figsize=(10,6),dpi=100)
plt.plot(w_range,emag_list,linestyle='solid',color='darkorange',label='e(w)')
plt.grid(True,alpha=0.6,linestyle='dashed',color='darkgray',which='both')
plt.legend(prop={'size':15})
plt.rcParams['legend.loc'] = 'upper right'
plt.title('e(w) - One Harmonic Oscillator')
plt.xlabel('w [rad/s]')
plt.ylabel('e(w)')
plt.show()

plt.figure(figsize=(10,6),dpi=100)
plt.plot(w_range,kw_list,linestyle='solid',color='limegreen',label='k(w)')
plt.plot(w_range,ei_list,linestyle='solid',color='hotpink',label='e_i(w)')
plt.grid(True,alpha=0.6,linestyle='dashed',color='darkgray',which='both')
plt.legend(prop={'size':15})
plt.rcParams['legend.loc'] = 'upper right'
plt.title('k(w) vs e_i(w) - One Harmonic Oscillator')
plt.xlabel('w [rad/s]')
plt.ylabel('k(w), ei(w)')
plt.show()
```
